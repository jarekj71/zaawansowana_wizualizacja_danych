# Bezpośrednia zmiana **rcparams** wpływa na wszystkie wykresy

cycler jest to funkcja, która udostępnia kolejne wartości dla danej własności wykresu. Są dwie metody zmiany parametru, bezpośrednia, jako parametr słownika, lub poprzez funkcję. Ta druga opcja pozawala zmienić naraz kilka parametrów.

```python
mpl.rcParams['scatter.marker']='x'
mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=['#FFa121', 'b', 'g', 'orange'])
plt.scatter(x,y) # Pierwszy element w cycler
plt.scatter(1,0.5) # Drugi element w cycler
```

Defniowanie koloru przy pomocy cyclera nie jest logiczne. Działa w zasadzie na wybrane elementy. Nie można definiować elementów pobocznych typu edgecolor

```python
print(dict([(k,v) for k,v in params.items() if k.startswith('scatter')]))
```

```python
plt.rcdefaults()
clc = plt.cycler(color=['#FFa1CC', 'tab:red', 'g', 'orange'])
plt.rc('axes',prop_cycle=clc)
plt.rc('scatter',marker='o',edgecolors='#FF41CC') # pierwszy kolor w cycler
for X,Y in zip(x,y):
    plt.scatter(X,Y)
```

# tymczasowa zmiana wybranych części stylu

Stosowane rzadko, parametry definiowane jako klucze słownika. Nie należy mylić z kontekstem dla seaborn

```python
with mpl.rc_context({'lines.markersize': 2, 'scatter.marker':'x',}): # lines.markersize definiuje wielkość dla scatter również, niekonsekwencja i bałagan w definicjach
    plt.scatter(x,y)
```

# Ręczne definiowanie stylu linii

styl linii definiujemy przez wartości (offset,(black,white)) 

Można utworzyć dowolny wzór

```python
plt.plot(x,y,ls=(0,(1,5)))
```

```python
vis = {'axes.facecolor':'#fff9d3',
 'axes.labelcolor': '#666666',
 'axes.prop_cycle': plt.cycler('linestyle', [(0,(1,1)),(0,(1,2)),(0,(2,1)),(0,(2,2)),(0,(3,5))]),
 'axes.linewidth': 1.0,
}

alt1 = {
'axes.linewidth': 2.0,
'lines.markersize' : 16,
'axes.titlesize': 'xx-large', # latex
'xtick.labelsize' : 16,
'ytick.labelsize' : 16
   
} 

alt2 = {
'lines.markersize' : 10,
'axes.labelsize': 'medium',
'axes.titlesize': 'large',
'xtick.labelsize' : 10,
'ytick.labelsize' : 10

} 
```

```python
plt.rcdefaults()
with mpl.rc_context({**vis,**alt2}): 
    plt.scatter(x,y)
    plt.xlabel("X oś")
    plt.ylabel("Y oś")
    plt.title("Tytuł")
    
```

# Style w Seaborn

Seaborn używa dwóch grup styli: **style** - odpowiedzialne za wizualizację i **context** (nie mylić z rc.context) odpowiedzialne za medium prezentacji, głównie wielkości elementów. Oba elemety razem tworzą **theme**: odpowiednio z parametrami style i context. osobno można również definiować palety kolorów i czcionki Polecenie sns.set_theme() bez żadnych parametrów resetuje styl do domyślnych

```python
import seaborn as sns
sns.set_theme()
sns.scatterplot(x=x, y=y) # default plot
```

```python
sns.set_theme(style='ticks',context='poster') #paper, notebook, talk, poster
sns.scatterplot(x=x, y=y) 
```

```python
sns.set_theme(style='ticks',context='paper') 
sns.scatterplot(x=x, y=y) 
```
