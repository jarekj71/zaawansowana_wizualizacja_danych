## Rozmieszczenie grafik na wykresie

Jest to bardzo podobny sposób, jak w przypadku wykresów kołowych, tyle że zamiast axes rozmieszczamy obiekty graifczne (artists)

1. Wczytanie obiektu graficznego do pamięci
2. Dodanie losowych koordynat

```python
import os
import matplotlib.image as mpimg
import matplotlib.offsetbox as mpoff 
import matplotlib.pyplot as plt
import numpy as np
```

## Generowanie grafiki z danych

```python
array = (np.random.rand(20,20)*100).astype(int)
trh = np.where(array<90,0,array-90)
```

```python
fig,ax = plt.subplots(figsize=(6,6))
ax.imshow(np.ma.masked_equal(trh,0),cmap='rainbow')
```

## Wstawianie grafik jako elementów wykresu

Grafiki można umieszczać w wykresie podobnie jak inne artists

```python
images = os.listdir("dane/cyfry")

imarray = {}
for image in images:
    imarray[image[:-4]] = mpimg.imread('dane/cyfry/'+image) #1

coords = np.random.rand(10,2) #2 
```

3. Utworzenie obiektu graficzngo klasy image
4. Utworzenie obiektu graficznego AnnotationBox zawierającego image
5. Dodanie obiektu jako artist

```python
fig,ax = plt.subplots(figsize=(12,12))
for coord,key in zip(coords,imarray.keys()):
    img = imarray[key]
    oimg = mpoff.OffsetImage(img,zoom=0.75,interpolation='none',cmap='gist_gray') #3
    ab = mpoff.AnnotationBbox(oimg, xy=coord, xybox=(0, 0), boxcoords="offset points") #4
    ax.add_artist(ab) #5
ax.set_xlim(-0.1,1.1)
ax.set_ylim(-0.1,1.1)
_ = ax.set_title("Images scattered on plot")
```
