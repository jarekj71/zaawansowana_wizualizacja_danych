## Skład chemiczny 

Zbiór obejmuje skład chemiczny próbek ceramiki z kilku stanowisk archeologicznycyh w okolicy Poznania. Wykonany wykres ma na celu pokazać niepodobieństwo (distance) pomiędzy tymi próbami oraz ich rozmiar i skład chemiczny. Pierwszym krokiem będzie wyliczenie niepodobieństwa. Alternatywnie, można wczytać plik chemia0.csv z już wyliczonym niepodobieństwem

```python
import scipy.spatial.distance as dst
from sklearn.manifold import MDS
chemia = pd.read_csv("dane/chemia.csv",index_col="Stanowisko")
dists = dst.pdist(chemia)
coords = MDS(dissimilarity='precomputed',normalized_stress='auto').fit_transform(dst.squareform(dists))
chemia['X'] = coords[:,0]
chemia['Y'] = coords[:,1]

```

```python
chemia = pd.read_csv("dane/chemia0.csv",index_col="Stanowisko") #jak nie ma MDS i SCIPY
```

To oznaczeń pierwiastków użyjemy kolorów z listy nazwanych kolorów

https://matplotlib.org/stable/gallery/color/named_colors.html

Możliwości wykresów kołowych opisane są w dokumentacji

https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

```python
colors = ["lavender","lightblue","bisque","indianred","floralwhite","dimgrey","indigo"]
```

1. Obliczenie parametru skalowania wykresów kołowych, tak aby wykresy zmieściły się w wykresie głównym
2. Operacje wykonujemy dla każdego wiersza osobno
3. Wielkość próbki reprezentowana jest przez wielkość wykresu kołowego
4. Każdy wykres kołowy wstaiwamy jako inset_axes, lokalizację wykresów ustawiamy na transData bo lokalizujemy je względem wspórzędnych skalowania wielowymiarowego. Nie ma potrzeby usuwania osi wykresu bo `.pie` domyślnie wyłącza osie
5. Tytuł wykresu przesuwamy w dół. Tu wartość y=0.6
6. Dodajemy wykres kołowy, który zwraca nam zarówno geometrię (patches), jak i przypisane im wartości (texts)
7. Patches i texts używamy do wykonania legendy

```python
fig,ax = plt.subplots(figsize=(10,10))

X_span = chemia.X.max()-chemia.X.min() #1
Y_span = chemia.Y.max()-chemia.Y.min()
scale = chemia.iloc[:,:-2].sum(axis=1).max()/max(X_span,Y_span)

for _,row in chemia.iterrows(): #2
    size = row[:-2].sum()/(scale*3) #3
    axx = ax.inset_axes([row.X-size/2,row.Y-size/2,size,size],transform=ax.transData) #4
    axx.set_title(row.name,y=0.6) #5
    patches, texts = axx.pie(row[:-2],colors=colors) #6

ax.set_xlim(chemia.X.min()*scale/3,chemia.X.max()*scale/3)
ax.set_ylim(chemia.Y.min()*scale/3,chemia.Y.max()*scale/3)
_ = ax.legend(patches,chemia.columns[:-2],loc='lower left') #7
```

## Rozmieszczenie grafik na wykresie

Jest to bardzo podobny sposób, jak w przypadku wykresów kołowych, tyle że zamiast axes rozmieszczamy obiekty graifczne (artists)

1. Wczytanie obiektu graficznego do pamięci
2. Dodanie losowych koordynat

```python
import os
import matplotlib.image as mpimg
import matplotlib.offsetbox as mpoff 
images = os.listdir("dane/cyfry")

imarray = {}
for image in images:
    imarray[image[:-4]] = mpimg.imread('dane/cyfry/'+image) #1

coords = np.random.rand(10,2) #2 
```

3. Utworzenie obiektu graficzngo klasy image
4. Utworzenie obiektu graficznego AnnotationBox zawierającego image
5. Dodanie obiektu jako artist

```python
fig,ax = plt.subplots(figsize=(12,12))
for coord,key in zip(coords,imarray.keys()):
    img = imarray[key]
    oimg = mpoff.OffsetImage(img,zoom=0.75,interpolation='none',cmap='gist_gray') #3
    ab = mpoff.AnnotationBbox(oimg, xy=coord, xybox=(0, 0), boxcoords="offset points") #4
    ax.add_artist(ab) #5
ax.set_xlim(-0.1,1.1)
ax.set_ylim(-0.1,1.1)
_ = ax.set_title("Images scattered on plot")
```
